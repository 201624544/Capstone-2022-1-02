```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, f1_score
import tensorflow as tf
import numpy
import seaborn as sns

csv1 = '/content/drive/MyDrive/final/test_labels.csv'
csv2 = '/content/drive/MyDrive/final/ttttt2.csv'


df1=pd.read_csv(csv1)
df2=pd.read_csv(csv2)

aValue=df1['MGMT_value']
pValue=df2['MGMT_value']

aValue=aValue.values
pValue=pValue.values

cmt=confusion_matrix(aValue, pValue)
print(cmt)
print("---------------------------------------")

accuracy=accuracy_score(aValue, pValue)
precision=precision_score(aValue, pValue)
recall=recall_score(aValue, pValue)

print('accuracy:', accuracy)
print('precision:', precision)
print('recall:', recall)

print("---------------------------------------")

f1=f1_score(aValue, pValue)
print('f1:', f1)
print("---------------------------------------")

from sklearn.metrics import classification_report
print(classification_report(aValue, pValue))

four = sns.heatmap(cmt, annot=True)
four.get_figure().savefig('/content/drive/MyDrive/final/ttt.png')

m = tf.keras.metrics.AUC(num_thresholds=3)
m.update_state(aValue, pValue)
print('AUC : ',m.result().numpy())
```
